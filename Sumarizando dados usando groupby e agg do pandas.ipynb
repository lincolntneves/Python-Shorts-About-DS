```python
# Importando pacotes
import pandas as pd
```

```python
# Criando Data Frame
data1 = {'id':[1,4,3,4,5],
         'gender':['f','m','f','f','m'],
         'ticket':[34.5,32.3,35.6,37.0,31.2],
         'count':[20,12,45,10,24]}
data2 = pd.DataFrame(data1)
data2
```

```python
# Sumarizando variáveis numéricas pelo id
data2.groupby(['id']).agg({'ticket': ['mean'],
                           'count':['sum']})
```

```python
# Sumarizando variáveis numéricas pelo id, mas usando duas estatísticas para uma das variáveis
data2.groupby(['id']).agg({'ticket': ['mean','median'],
                           'count':['sum']})
```

```python
# Sumarizando variáveis numéricas pelo id, igual ao anterior e
# contando o volume de id (linhas com id igual)
data2.groupby(['id']).agg({'id': ['count'],
                           'ticket': ['mean','median'],
                           'count':['sum']})
```

```python
# Sumarizando variáveis numéricas pelo genero, usando medidas igual ao anterior
data2.groupby(['gender']).agg({'id': ['count'],
                               'ticket': ['mean','median'],
                               'count':['sum']})
```

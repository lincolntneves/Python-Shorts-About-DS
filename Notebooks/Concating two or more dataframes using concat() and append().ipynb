# Concating two or more dataframes using concat() and append()

The objective of this notebook is only to show up how to use concat() and append() to union two or more dataframes.

```python
# Reading packages
import pandas as pd
```

```python
# Building the first dataframe
dic1 = {'id':[1,4,3,4,5],
         'gender':['f','m','f','f','m'],
         'ticket':[34.5,32.3,35.6,37.0,31.2],
         'count':[20,12,45,10,24]}
data1 = pd.DataFrame(dic1)
data1
```

```python
# Building the second dataframe
dic2 = {'id':[3,3,3,5,1],
         'gender':['m','m','f','m','f'],
         'ticket':[21.5,45.3,20.6,19.0,28.2],
         'count':[48,18,98,32,43]}
data2 = pd.DataFrame(dic2)
data2
```

```python
# Building an empty list
list_data = []
```

```python
# Appending dataframes on the list
list_data.append(data1)
list_data.append(data2)
```

```python
# Concating all dataframes
data3 = pd.concat(list_data,ignore_index = True)
data3
```
